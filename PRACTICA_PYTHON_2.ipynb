# <font face="gotham" color="black"> **4. Ejemplo Introductorio**</font>

````python
plt.figure(figsize=(10, 6))
sns.histplot(datos1['fallecidos'], bins=10, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribución de Fallecidos en Batallas')
plt.xlabel('Número de Fallecidos')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
````
